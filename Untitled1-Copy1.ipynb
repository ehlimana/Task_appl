```python

# Path do google chrom webdriver

PATH = "C:\\Users\\Ehlimana\\Desktop\\chromedriver.exe"
driver = webdriver.Chrome(PATH)
article_list = []
window_before = driver.window_handles[0]
# url stranice koju hocemo scrape-ati
URL_FOR_SCRAPING = 'https://feeds.finance.yahoo.com/rss/2.0/headline?s=AAPL&region=US&lang=en-US'

driver.get(URL_FOR_SCRAPING)

next1 = driver.find_elements_by_tag_name('item')
for a in next1 :
        print('Starting the scraping tool')
        title = a.find_element_by_tag_name ("title")
        print(title.get_attribute("textContent"))
        
        description = a.find_element_by_xpath ('description')
        print(description.get_attribute("textContent"))
      
        link = a.find_element_by_tag_name ("link")
        print(link.get_attribute("textContent"))
        
        published = a.find_element_by_xpath ('pubDate')
        print(published.get_attribute("textContent"))
       
        article = {
                'title': title,
                'description': link,
                'link': link,
                'published': published  }
   
        article_list.append(article)
        print('Finished scraping the articles')
    
            # after the loop, dump my saved objects into a .txt file

```
